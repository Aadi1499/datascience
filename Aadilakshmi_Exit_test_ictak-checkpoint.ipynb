# Data preprocessing

Handling missing values

```python
df.isnull().sum()
```

```python
#find the percentage of missing values
features_with_na = [features for features in data.columns if data[features].isnull().sum()>=1]

for feature in features_with_na:
    print(feature,np.round(data[feature].isnull().mean(),3), ' % missing values') 
```

```python
for feature in ['gender', 'coaching' ,'strong_foot']:
    df[feature]=df[feature].fillna(df[feature].mode()[0])
```

```python
for feature  in ['weight','ball_controlling_skills','jumping_skills','penalties_conversion_rate','mental_strength','shot_accuracy','behaviour_rating','matches_played','fitness_rating','years_of_experience']:
    df[feature ]=df[feature].fillna(round(df[feature ].median(),1))
```

```python
df.isnull().sum()
```

Outlier handling

```python
for feature in continuous_feature:
    if 0 in df[feature].unique():
        pass
    else:
        df[feature] = np.log(df[feature])
        df.boxplot(column = feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

The outliers are found in all the features except 'years_of_experience'

```python
outliers=outliers=['age', 'height','weight','ball_controlling_skills','body_reflexes','body_balance','jumping_skills','penalties_conversion_rate','goalkeeping_skills']

 #Outliers are seen above and below the limit

for feature in outliers:
  Q1 = df[feature].quantile(0.25)
  Q2 = df[feature].quantile(0.50)
  Q3 = df[feature].quantile(0.75)
  IQR=Q3-Q1
  upper_limit=Q3+1.5*IQR
  lower_limit=Q1-1.5*IQR
  df[feature]=np.where(df[feature]>upper_limit,upper_limit,np.where(df[feature]<lower_limit,lower_limit,df[feature]))
```

Encoding

```python
#the categorcial coloumns are encoded
df_1 = df[['gender','strong_foot','dedication_level','coaching']]
df_encoded = pd.get_dummies(df_1,drop_first = True)
```

```python
df_encoded.head()
```

```python
df.head()
```

```python
data = pd.concat([df[continuous_feature],df_encoded,df[discrete_feature]],axis=1)
```

# Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data[continuous_feature] = scaler.fit_transform(data[continuous_feature])
data[continuous_feature] =pd.DataFrame(data[continuous_feature],columns=data[continuous_feature].columns)
data[continuous_feature].describe()
```

```python
data.head()
```

Train-Test splitting of final dataset

```python
x= data.drop('selection',axis=1)
y= data['selection']
```

```python
#split into train-test sets
from sklearn.model_selection import train_test_split
from sklearn import linear_model
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.25)
```

# Modeling

```python
#decision tree model
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
dt_model = DecisionTreeClassifier()
dt_model.fit(x_train, y_train)
dt_pred  = dt_model.predict(x_test)

print('Accuracy: ',accuracy_score(y_test,dt_pred))
print('Precision is: ',precision_score(y_test,dt_pred,average='micro'))
print('Recall Score: ',recall_score(y_test,dt_pred,average='micro'))
print('f1 Score: ',f1_score(y_test,dt_pred,average='micro'))
```

```python
#support vector machine model
from sklearn.svm import SVC
svm_linear = SVC(kernel="linear")
svm_linear.fit(x_train, y_train)
svm_pred  = svm_linear.predict(x_test)

print('Accuracy: ',accuracy_score(y_test,svm_pred))
print('Precision: ',precision_score(y_test,svm_pred,average='micro'))
print('Recall Score: ',recall_score(y_test,svm_pred,average='micro'))
print('f1 Score: ',f1_score(y_test,svm_pred,average='micro'))
```

```python
#random forest classification
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 42)
rf.fit(x_train, y_train)
rf_pred = rf.predict(x_test)

print('Accuracy: ',accuracy_score(y_test,rf_pred))
print('Precision: ',precision_score(y_test,rf_pred,average='micro'))
print('Recall Score: ',recall_score(y_test,rf_pred))
print('f1 Score: ',f1_score(y_test,rf_pred))
```

The random forest classifier shows the best accuracy and precision of 0.84 and recall score of 0.87
